```python
code = []
file = open("code.txt", 'r')
for line in file:
    code += [line.strip()]

file.close()
```

```python
# global dict/set
operator_set = set({"sh", "op", "li", "b", "lb", "sb", "mov", "par"})
o_to_m = {"sh" : "000", "op" : "001", "li" : "010", "b" : "011", "lb" : "101", "sb" : "110", "mov" : "100", "par" : "111"}
r_to_m = {"$b" : "000", "$k" : "001", "$r1": "010", "$r2" :"011", "$r3" : "100", "$r4" : "101", "$r5" : "110", "$r6" : "111"}
```

```python
# change aseembly code into machine code
output = []
out_file = open("output.txt", 'w')
for line in code:
    m = ""
    k = line.lower().split()
    
    if len(k) == 0: continue
    if k[0] not in operator_set: continue
    cmd = k[0]
    sec = k[1][:-1]

    m += o_to_m[k[0]]
    m += r_to_m[sec]
    if k[0] == "sh":
        m += k[2][-3:]
    elif k[0] == "op":
        m += k[2][-3:]
    elif k[0] == "li":
        m += k[2][-3:]
    elif k[0] == "b":
        m += k[2][-3:]
    elif k[0] == "lb":
        m += r_to_m[k[2]]
    elif k[0] == "sb":
        m += r_to_m[k[2]]
    elif k[0] == "mov":
        m += r_to_m[k[2]]
    elif k[0] == "par":
        m += r_to_m[k[2]]
    else:
        print("error")
    out_file.write(m+"\n")
out_file.close()
```
